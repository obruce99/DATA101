Part 1: Working with Numpy

Step 1: Use for loop to input names and scores. 

```python
import numpy as np

names = []
grades = np.zeros((10,5))

for x in range(10):
    name = input(f"Enter name for student {x+1}: ")
    names.append(name)
    print(f"Enter five scores for {name}: ")
    for y in range(5):
        grades[x,y] = float(input(f"  Assessment {y+1}:  "))
```

Step 2: Create Weight Vector

```python
weights = np.array([.15,.15,.20,.20,.30])
```

Step 3: Calculate Final Number Scores

```python
finalscores = np.dot(grades,weights)
```

Step 4: Define function to calculate letter grades. 

```python
def lettergrade(score):
    if score >= 90: return 'A'
    elif score >= 80: return'B'
    elif score >=70: return 'C'
    elif score >=60: return 'D'
    else : return 'F'
```

Step 5: Print the final results with Name, Number Score, and Letter Grade. 

```python
print(f"{'Student Name':<15} | {'Score':<6}  | {'Grade'}\n")

for x in range(10):
    letter = lettergrade(finalscores[x])
    print(f"{names[x]:<15} | {finalscores[x]:>6.2f} | {letter}")
```
